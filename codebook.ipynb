```python
#limpiar a la primera fila que tenga fecha 

def clean(sheet):
    raw_df = pd.read_excel(file, sheet_name=sheet)
    #encontrar el indice del encabezado
    header_idx = raw_df[raw_df.apply(lambda row: row.astype(str).str.contains("Fecha").any(), axis=1)].index[0]
    df= pd.read_excel(file, sheet_name=sheet, skiprows=header_idx )
    return df

importacion_clean_df = clean('IMPORTACION')
consumo_clean_df = clean('CONSUMO')

```

# Generar codebook

```python
from pandas.api.types import is_numeric_dtype, is_datetime64_any_dtype

def generate_codebook_completo(df):
    records = []
    for col in df.columns:
        series = df[col]
        dtype = series.dtype
        missing_count = series.isnull().sum()

        # Valores posibles
        if series.apply(lambda x: isinstance(x, (dict, list))).any():
            safe_series = series.dropna().apply(lambda x: str(x))
        else:
            safe_series = series.dropna()

        uniques = safe_series.unique()
        if len(uniques) <= 20:
            posibles = uniques.tolist()
        else:
            posibles = f"{len(uniques)} valores unicos"

        # Estadisticas basicas
        if is_numeric_dtype(series):
            stats = series.describe().to_dict()
        elif is_datetime64_any_dtype(series):
            stats = {"min": series.min(), "max": series.max()}
        else:
            top = safe_series.mode().iloc[0] if not safe_series.mode().empty else None
            freq = safe_series.value_counts().iloc[0] if not safe_series.value_counts().empty else None
            stats = {"unique": safe_series.nunique(), "top": top, "freq": freq}

        records.append({
            "nombre": col,
            "descripcion": "",  #llenar 
            "tipo de dato": str(dtype),
            "unidades": "Barriles (42 galones)" if "Gasolina" in col or "Diesel" in col else "",
            "valores posibles": posibles,
            "valores especiales/faltantes": missing_count,
            "estadisticas basicas": stats,
            "fuente original": "Estadisticas historicas de comercializacion",
            "transformaciones": "dropna, filtrado columnas, conversiÃ³n a datetime" if col == "Fecha" or "Gasolina" in col else "",
            "notas": ""
        })
    return pd.DataFrame(records)

# codebooks para importacion y consumo
codebook_importacion_final = generate_codebook_completo(importacion_clean_df)
codebook_consumo_final = generate_codebook_completo(consumo_clean_df)


```

```python


#llenar la descripcion
codebook_importacion_final['descripcion'] = [
    "Fecha (aÃ±o y mes) en que se registrÃ³ la importaciÃ³n del combustible.",
    "Volumen importado de gasolina regular en el mes correspondiente.",
    "Volumen importado de gasolina superior en el mes correspondiente.",
    "Volumen importado de diÃ©sel con bajo contenido de azufre (normas ambientales).",
    "Volumen importado de diÃ©sel con ultra bajo contenido de azufre (normas mÃ¡s estrictas).",
    "Volumen importado de diÃ©sel con alto contenido de azufre (menos regulado).",
    "Mes (1-12) en que se realizÃ³ la importaciÃ³n. Generado a partir de la columna Fecha.",
    "AÃ±o en que se realizÃ³ la importaciÃ³n. Generado a partir de la columna Fecha.",

]
codebook_consumo_final['descripcion'] = [
    "Fecha (aÃ±o y mes) en que se registrÃ³ el consumo del combustible.",
    "Volumen de gasolina regular consumido en el mercado nacional en el mes correspondiente.",
    "Volumen de gasolina superior consumido en el mercado nacional en el mes correspondiente.",
    "Volumen consumido de diÃ©sel con bajo contenido de azufre.",
    "Volumen consumido de diÃ©sel con ultra bajo contenido de azufre.",
    "Volumen consumido de diÃ©sel con alto contenido de azufre",
    "Mes (1-12) en que se realizÃ³ el consumo. Generado a partir de la columna Fecha.",
    "AÃ±o en que se realizÃ³ el consumo. Generado a partir de la columna Fecha.",
]

#notas
codebook_importacion_final['notas'] = "Datos de importaciÃ³n y consumo de combustibles lÃ­quidos en Guatemala, se encuentra en el dataset de importaciones"
codebook_consumo_final['notas'] = "Datos de importaciÃ³n y consumo de combustibles lÃ­quidos en Guatemala, se encuentra en el dataset de consumo"



codebook_total = pd.concat([codebook_importacion_final, codebook_consumo_final], ignore_index=True)

#exportarlo a csv
codebook_total.to_csv('codebook_total.csv', index=False)
codebook_total.to_excel('codebook_total.xlsx', index=False)
```
